```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
classdata_df = pd.read_csv('CSCI_Fall_2024_survey.csv')
class_labels = LabelEncoder()
classdata_df.head()
```

```python
classdata_df["Food"]=np.where(classdata_df["What is your favorite type of food?"]=="Japanese","J",
                              np.where(classdata_df["What is your favorite type of food?"]=="Mexican","M",
                                       np.where(classdata_df["What is your favorite type of food?"]=="Italian","It",
                                            np.where(classdata_df["What is your favorite type of food?"]=="American","A",
                                                     np.where(classdata_df["What is your favorite type of food?"]=="korean","K",
                                                              np.where(classdata_df["What is your favorite type of food?"]=="Indian","I",
                                                                       np.where(classdata_df["What is your favorite type of food?"]=="Chinese","C","all of the above")))))))
classdata_df.head(10)
```

```python
classdata_df["Major"]=np.where(classdata_df["What is your major?"]=="computer science","CS",
                               np.where(classdata_df["What is your major?"]=="Computer Science","CS",
                                        np.where(classdata_df["What is your major?"]=="Data Science","DS",
                                                 np.where(classdata_df["What is your major?"]=="CS/Business","CS/BU",
                                                          np.where(classdata_df["What is your major?"]=="cs/finance","CS/FI",
                                                                   np.where(classdata_df["What is your major?"]=="Biology","BIO",
                                                                            np.where(classdata_df["What is your major?"]=="Economics","ECON","TBD")))))))
classdata_df.head(10)
                               
```

```python
classdata_df["Age"]=np.where(classdata_df["How old are you?"]==18,1,
                             np.where(classdata_df["How old are you?"]==19,2,
                                      np.where(classdata_df["How old are you?"]==20,3,
                                               np.where(classdata_df["How old are you?"]==21,4,0))))
classdata_df.head()
                             
```

```python
classdata_df["Pets"]=np.where(classdata_df["How many pets and of which kind do you have?"]=="0","zero pets",
                              np.where(classdata_df["How many pets and of which kind do you have?"]=="1","one pet",
                                       np.where(classdata_df["How many pets and of which kind do you have?"]=="I have two dogs and they are both at home","two pets",
                                                np.where(classdata_df["How many pets and of which kind do you have?"]=="no pets","zero pets",
                                                         np.where(classdata_df["How many pets and of which kind do you have?"]=="1 dog and 1 cat","two pets",
                                                                  np.where(classdata_df["How many pets and of which kind do you have?"]=="2 cats, one dog","three pets",
                                                                           np.where(classdata_df["How many pets and of which kind do you have?"]==2,"two pets",
                                                                                    np.where(classdata_df["How many pets and of which kind do you have?"]=="1 Cat; 6 Dogs (Over two households)","seven pets",
                                                                                              np.where(classdata_df["How many pets and of which kind do you have?"]=="had 8 dogs, not anymore","eight pets",
                                                                                                       np.where(classdata_df["How many pets and of which kind do you have?"]=="three dogs","three pets",
                                                                                                                np.where(classdata_df["How many pets and of which kind do you have?"]=="1, Dog","one pet",
                                                                                                                         np.where(classdata_df["How many pets and of which kind do you have?"]=="no pets!","zero pets","zero pets"))))))))))))
classdata_df.head(10)
                                                                                             
                              
```

```python
classdata_df["Origin"]=np.where(classdata_df["What state/country were you born in?"]=="Mexico","MEX",
                                np.where(classdata_df["What state/country were you born in?"]=="Frankfurt, Germany","GER",
                                         np.where(classdata_df["What state/country were you born in?"]=="India ","IND",
                                                  np.where(classdata_df["What state/country were you born in?"]=="Georgia(the country not the state)","GEO",
                                                           np.where(classdata_df["What state/country were you born in?"]=="I was born in Chandigarh, India!", "IND",
                                                                    np.where(classdata_df["What state/country were you born in?"]=="People's Republic of China","CHI",
                                                                             np.where(classdata_df["What state/country were you born in?"]=="Cameroon","CAM",
                                                                                      np.where(classdata_df["What state/country were you born in?"]=="Kazakhstah","KAZ","USA"))))))))
classdata_df.head(10)
                                
```

```python
class_labels = LabelEncoder()
classdata_df['Major'] = class_labels.fit_transform(classdata_df['Major'].values)
classdata_df
```

```python
origin_count= classdata_df["Origin"].value_counts()

origin_count.plot(kind="bar")
plt.xlabel('Origin')
plt.ylabel('Count')
plt.title('Histogram of Country of Birth - CSCI 1070 Students')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()
```

The histogram above shows the count of all the students born in each country in the CSCI 1070 class

```python
agg_food = classdata_df.groupby("Food").count()
agg_food
```

```python
colors = ['pink','blue']
agg_food.plot(kind="pie", y = "Major", colors=colors)
```

The pie chart above shows the numerical comparison between CSCI 1070 students favorite food and major.

```python
plt.plot(classdata_df["Pets"], classdata_df["Age"],'.')
```

The scatterplot above shows the correlation between the number of pets CSCI 1070 students have and the allocated value for each age(17=0, 18=1,19=2,etc.)

```python
pet_count= classdata_df["Pets"].value_counts()

pet_count.plot(kind="bar")
plt.xlabel('Number of Pets')
plt.ylabel('Student Count')
plt.title('Histogram of Pet Count among CSCI 1070 Students')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()
```

The histogram above shows the student count for each amount of pets among students in the CSCI 1070 class.

```python
agg_age = classdata_df.groupby("Age").count()
agg_age
```

```python
colors = ['pink','blue']
agg_age.plot(kind="pie", y = "Major", colors=colors)
```

The pie chart above shows the correlation among students ages' and their majors. It can be seen from the chart that most students in the CSCI 1070 class are 18 years old and computer science majors.

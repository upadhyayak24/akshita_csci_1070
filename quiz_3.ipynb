```python
import pandas as pd
belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df.tail()
```

```python
groupdf = belize_df.groupby('neighbourhood')
```

```python
pricedf = groupdf[["price"]].mean().reset_index()
pricedf
```

```python
reviewdf = groupdf[["number_of_reviews"]].sum().reset_index()
reviewdf
```

```python
availabilitydf = groupdf[["availability_365"]].max().reset_index()
availabilitydf
```

```python
inner_merge_df = pd.merge(pricedf, reviewdf, on="neighbourhood")
inner_merge_df
```

```python
merge_df = pd.merge(inner_merge_df, availabilitydf, on="neighbourhood")
merge_df
```

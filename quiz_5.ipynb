```python
#2
import numpy as np
import pandas as pd

data_df = pd.read_csv("JEOPARDY_DATA.csv")
data_df.head(10)
```

```python
data_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

class_labels = LabelEncoder()

data_df[' Round'] = class_labels.fit_transform(data_df[' Round'].values)
data_df = pd.get_dummies(data_df[[' Round']])
data_df
```

```python
data_df[' Value'] = class_labels.fit_transform(data_df[' Value'].values)
data_df[' Value'] = pd.to_numeric(data_df[' Value'])
data_df[' Value'] = data_df[' Value'].astype(float)
print(data_df.dtypes)
```

```python
import nltk
import matplotlib.pyplot as plt
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Download necessary NLTK resources
nltk.download('gutenberg', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('stopwords', quiet=True)

# Read Paradise Lost from Gutenberg Corpus
paradise_lost_text = gutenberg.raw('milton-paradise.txt')

# Tokenize and convert to lowercase
tokens = word_tokenize(paradise_lost_text.lower())

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Lemmatize tokens and filter out stop words and non-meaningful tokens
def is_meaningful_token(token):
    # Exclude stop words, single characters, and very short tokens
    return (token not in stop_words and 
            len(token) > 2 and 
            token.isalpha())

# Lemmatize meaningful tokens
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if is_meaningful_token(token)]

# Manual word frequency counting
word_freq = {}
for word in lemmatized_tokens:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

# Sort words by frequency in descending order
sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Get top 20 words
top_20_words = sorted_words[:20]

# Create bar plot
plt.figure(figsize=(15, 7))
words, counts = zip(*top_20_words)
plt.bar(words, counts)
plt.title('Top 20 Words in Paradise Lost (Lemmatized)', fontsize=15)
plt.xlabel('Words', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()

# Print out the top 20 words and their counts for reference
print("Top 20 Words and Their Frequencies:")
for word, count in top_20_words:
    print(f"{word}: {count}")
```

```python
import nltk
import re
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer

# Download necessary NLTK resources
nltk.download('gutenberg', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('vader_lexicon', quiet=True)

# Read Paradise Lost from Gutenberg Corpus
paradise_lost_text = gutenberg.raw('milton-paradise.txt')

# Clean the text (remove line breaks, extra whitespace)
def clean_text(text):
    # Remove multiple whitespaces
    text = re.sub(r'\s+', ' ', text)
    # Remove newline characters
    text = text.replace('\n', ' ')
    return text

cleaned_text = clean_text(paradise_lost_text)

# Tokenize into sentences
sentences = sent_tokenize(cleaned_text)

# Initialize Vader Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def analyze_sentiment(sentence):
    scores = sia.polarity_scores(sentence)
    return {
        'sentence': sentence,
        'compound': scores['compound'],
        'pos': scores['pos'],
        'neu': scores['neu'],
        'neg': scores['neg']
    }

# Analyze sentiments of sentences
sentence_sentiments = [analyze_sentiment(sent) for sent in sentences]

# Sort sentences by different sentiment metrics
most_negative = sorted(sentence_sentiments, key=lambda x: x['compound'])[:5]
most_positive = sorted(sentence_sentiments, key=lambda x: x['compound'], reverse=True)[:5]
most_neutral = sorted(sentence_sentiments, key=lambda x: x['neu'], reverse=True)[:5]

# Print results
print("5 Most Negative Sentences:")
for sent in most_negative:
    print(f"Sentiment Score: {sent['compound']:.4f}")
    print(f"Sentence: {sent['sentence']}\n")

print("\n5 Most Positive Sentences:")
for sent in most_positive:
    print(f"Sentiment Score: {sent['compound']:.4f}")
    print(f"Sentence: {sent['sentence']}\n")

print("\n5 Most Neutral Sentences:")
for sent in most_neutral:
    print(f"Neutrality Score: {sent['neu']:.4f}")
    print(f"Sentence: {sent['sentence']}\n")

# Optional: Create a summary of overall sentiment
total_sentences = len(sentence_sentiments)
positive_count = sum(1 for s in sentence_sentiments if s['compound'] > 0)
negative_count = sum(1 for s in sentence_sentiments if s['compound'] < 0)
neutral_count = sum(1 for s in sentence_sentiments if s['compound'] == 0)

print("\nSentiment Distribution:")
print(f"Total Sentences: {total_sentences}")
print(f"Positive Sentences: {positive_count} ({positive_count/total_sentences*100:.2f}%)")
print(f"Negative Sentences: {negative_count} ({negative_count/total_sentences*100:.2f}%)")
print(f"Neutral Sentences: {neutral_count} ({neutral_count/total_sentences*100:.2f}%)")
```

The scores above for both the Sentiment Analysis and the Lemmitization look correct. The words for the Lemmetized score are not stop words or words with one letter so the score looks fairly accurate. The sentiment score also look accurate because when you read the sentences themselves, the function picked the correct positive and negative sentences.
